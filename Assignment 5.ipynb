```python
import matplotlib.pyplot as plt 
import random

# vertices of an equilateral triangle (all three sides are equal in size and angles) 

vertices = [(0,0), (1, 0), (0.5,0.866)]

# a curated function to determine if a point exists within the triangle or not 

def isInside (x1, y1, x2, y2, x3, y3, x, y):
    def area (x1, y1, x2, y2, x3, y3):
        return abs((x1*(y2 - y3) + x2*(y3 - y1) + x3*(y1 - y2)) / 2.0 )

x1, y1 = 0, 0
x2, y2 = 1, 0
x3, y3 = 0.5, 0.866

A = area(x1, y1, x2, y2, x3, y3)
A1 = area(x, y, x2, y2, x3, y3)
A2 = area(x1, y1, x, y, x3, y3)
A3 = area(x1, y1, x2, y2, x, y)

return A == A1 + A2 + A3

# an input value is nesscary for the code to run, so the user must choose a seed point for the specific function

while True:
    try:
        seed_x = float(input("Enter the x-coordinate of the seed point: "))
        seed_y = float(input("Enter the y-coordinate of the seed point: "))
        if isInside (0, 0, 1, 0, 0.5, 0.866, seed_x, seed_y):
            print ("Valid seed point entered.")
            break
        else:
            print("The point is not inside the triangle. Please try again.")
    except ValueError:
        print("Invalid input. Please enter numerical values")

# this part of the code is where the user can start with seed points, otherwise known as the starting point
seed = (seed_x, seed_y)

# creating a list of all the numerical seed points will help store for later use when running the code
# a starting seed point can now be run from this moment on

points = {seed}

# from here, a number of steps will follow 

while True: 
    try:
        num_steps = int(input("Enter the number of steps: "))
        if num_steps > 0:
            print (f"Number of steps set to [num_steps].")
            break
        else:
            print ("Please enter a positive integer")
    except ValueError:
        print("Invalid input. Please enter a positive integer.")

for i in range(num_steps):
    # choosing a random vertex value for to cater the function towards the main point 
    next_vertex = vertices [random.randint(0, 2)]

    #to create the next point in the function, move the last point to the midpoint 
    next_point = ((points[-1][0] + next_vertex [0]) / 2, (points [-1][1] + next_vertex[1]) / 2)

    #add the new, coded point to the collected list of points to finish
    points.append(next_point)

# to add the finishing touch, plot the solution using the created function

plt.scatter([p[0] for p in points], [p[1] for p in points], s=0.1)
plt.show()

print(points)

        
```
