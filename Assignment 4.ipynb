```python
import numpy as np
import matplotlib.pyplot as plt

def gradient_descent (f, learning_rate, initial_point):
    def deriv(f,base_point): #estimiates derivative of functiion f at base_point
        return (f(base_point + 1e-5) - f(base_point)) / 1e-5

    x_coords = [initial_point] #list where x_n's are stored
    y_coords = [f(initial_point)] #list where y_n's are stored

    for _ in range (10000):  #Adjust the number of iterations as needed
        gradient = deriv(f, x_coords[-1])
        update = -learning_rate * gradient #the learning rate is multiplied by the derivative of the last x_n, which altogether is added to the last x_n to determine the next x_n
        x_coords.append(x_coords[-1] + update)
        y_coords.append(f(x_coords[-1]))

    plot_range = np.linspace(min(x_coords)- 0.5, max (x_coords) + 0.5, 1000) #appealing plot range
    function_range = [f(i) for i in plot_range]
    plt.plot(plot_range, function_range)
    plt.plot(x_coords, y_coords, marker='o', color = 'r')
    plt.xlabel('x')
    plt.ylabel('f(X)')
    plt.title('Gradient Descent Optimization')
    plt.show()

    return round(x_coords[-1], 3), round (y_coords[-1], 3) #returns fial x_n and y_n, rounded to 3 decimal places 

    #enter function to be optimized at this point 
def f(x):
   return  x**2

learning_rate = 0.1 #size of step/jump back and forth for each iteration 
initial_point = 1.0 #point from which gradient descent starts is set

result = gradient_descent(f, learning_rate, initial_point)
print ("Local Minima:", result) #prints local coordinate pair 
    
```
